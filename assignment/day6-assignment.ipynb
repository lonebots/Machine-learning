# ***random_forest_regression***

```
#libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
#importing the dataset
data=pd.read_csv('/content/drive/MyDrive/ML/TinkerWeek.py/dataset/Position_Salaries.csv')
x=data.iloc[:,1:-1].values
y=data.iloc[:,-1].values
```

```
#splitting the dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,_y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```
#getting the regressor model ready
#model_v1
#n_estimators=100(standard/default)
from sklearn.ensemble import RandomForestRegressor
rfr_model_1=RandomForestRegressor(random_state=0)
rfr_model_1.fit(x_train,y_train)
```

```
x_grid = np.arange(min(x), max(x), 0.01)
x_grid = x_grid.reshape((len(x_grid), 1))
plt.scatter(x, y, color = 'red')
plt.plot(x_grid, rfr_model_1.predict(x_grid), color = 'blue')
plt.title('Truth or Bluff (Random Forest Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```
#getting the regressor model ready
#model_v2
#n_estimators=500(hyper_parameter_tuning)
from sklearn.ensemble import RandomForestRegressor
rfr_model_2=RandomForestRegressor(n_estimators=500,random_state=0)
rfr_model_2.fit(x_train,y_train)
```

```
x_grid = np.arange(min(x), max(x), 0.01)
x_grid = x_grid.reshape((len(x_grid), 1))
plt.scatter(x, y, color = 'red')
plt.plot(x_grid, rfr_model_2.predict(x_grid), color = 'blue')
plt.title('Truth or Bluff (Random Forest Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```
#getting the regressor model ready
#model_v3
#n_estimators=5(hyper_parameter_tuning)
from sklearn.ensemble import RandomForestRegressor
rfr_model_3=RandomForestRegressor(n_estimators=5,random_state=0)
rfr_model_3.fit(x_train,y_train)
```

```
x_grid = np.arange(min(x), max(x), 0.01)
x_grid = x_grid.reshape((len(x_grid), 1))
plt.scatter(x, y, color = 'red')
plt.plot(x_grid, rfr_model_3.predict(x_grid), color = 'blue')
plt.title('Truth or Bluff (Random Forest Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

# ***random_forest_classification***

```
#libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
#importing the dataset
data=pd.read_csv('/content/drive/MyDrive/ML/TinkerWeek.py/dataset/Social_Network_Ads.csv')
x=data.iloc[:,0:-1].values
y=data.iloc[:,-1].values
```

```
#splitting the dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```
#scalling the data
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```
#getting the model ready
#model_v1
#n_estimators=100(default/standard)
from sklearn.ensemble import RandomForestClassifier
rfc_model_1=RandomForestClassifier(criterion='entropy',random_state=0)
rfc_model_1.fit(x_train,y_train)
```

```
from matplotlib.colors import ListedColormap
x_set, y_set = sc.inverse_transform(x_test),y_test
x1, x2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 10, stop = x_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = x_set[:, 1].min() - 1000, stop = x_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(x1, x2, rfc_model_1.predict(sc.transform(np.array([x1.ravel(), x2.ravel()]).T)).reshape(x1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Random Forest Classification (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```
#getting the model ready
#model_v2
#n_estimators=500(hyper_parameter_tuning)
from sklearn.ensemble import RandomForestClassifier
rfc_model_2=RandomForestClassifier(n_estimators=300,criterion='entropy',random_state=0)
rfc_model_2.fit(x_train,y_train)
```

```
from matplotlib.colors import ListedColormap
x_set, y_set = sc.inverse_transform(x_test),y_test
x1, x2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 10, stop = x_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = x_set[:, 1].min() - 1000, stop = x_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(x1, x2, rfc_model_2.predict(sc.transform(np.array([x1.ravel(), x2.ravel()]).T)).reshape(x1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Random Forest Classification (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```
#getting the model ready
#model_v3
#n_estimators=35(hyper_parameter_tuning)
from sklearn.ensemble import RandomForestClassifier
rfc_model_3=RandomForestClassifier(n_estimators=5,criterion='entropy',random_state=0)
rfc_model_3.fit(x_train,y_train)
```
